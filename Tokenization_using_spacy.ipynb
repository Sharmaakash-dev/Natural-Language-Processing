```python
# import the toolkit and the full porter stemmer lirary
import nltk
from nltk.stem.porter import PorterStemmer as p_stemmer
```

```python
words=['run','runner','running','ran','runs','easily','fairly']
```

```python
# p_stemmer = p_stemmer()
for word in words:
  print(word+'---->'+p_stemmer().stem(word))
```

```python
from nltk.stem.snowball import SnowballStemmer
```

```python
s_stemmer=SnowballStemmer(language='english')
```

```python
for word in words:
  print(word+'---->'+s_stemmer.stem(word))
```

#**LEMMATIZATION**

```python
doc= nlp(u'I am a runner running in a race because I love to run since I ran today')
```

```python
for token in doc:
  print(token.text,'\t',token.pos_,'\t',token.lemma,'\t',token.lemma_)
```

```python
for token in doc:
  print(token.text,'\t',token.pos_,'\t',token.lemma_)
```

```python
for word in doc:
  print(f"{word.text:{12}} {word.pos_:{6}} {word.lemma_}")
```

```python
for word in words:
  print(f"{word.lemma_}")
```

```python
def show_lemmas(text):
  for token in text:
    print(f'{token.text:{12}} {token.pos_:{6}} {token.lemma_}')
```

```python

doc2=nlp(u'I saw ten mice today!')
```

```python
show_lemmas(doc2)
```

```python
doc3=nlp(u"I am meeting him tomorrow at the meeting")
show_lemmas(doc3)
```

```python
doc4=nlp(u"That's an enormous automobile")
show_lemmas(doc4)
```
