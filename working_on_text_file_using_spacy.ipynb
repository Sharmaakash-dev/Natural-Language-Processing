#**Sub-Word based tokeniztion**

```python
str = "I'm learning tokenization."
```

```python
str1 = "I'm learning tokenization"
```

```python
new_str = str.split()
```

```python
list3=[]
for word in str.split():
  if word[-3:] =='ing':
    list3.append(word.partition('ing'))
  elif word[-7:] =='ization':
    list3.append(word.partition('ization'))
  else:
    list3.append(word)
print(list3)
```

```python
list3=[]
for word in str1.split():
  if word[-3:] =='ing':
    list3.append(word.partition('ing'))
  elif word[-7:] =='ization':
    list3.append(word.partition('ization')) # the full stop at the end makes a diff tokenization
  else:
    list3.append(word)
print(list3)
```

```python
list4= []
for element in list3:
  if type(element) == tuple:
    list4.append(element[0])
    list4.append(element[1])
```
