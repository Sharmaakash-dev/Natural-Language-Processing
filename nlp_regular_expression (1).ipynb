#**The Wildcard Character**

```python
re.findall(r".at","the cat in the hat sat here") # "." is used to find wildcard characters
```

```python
re.findall(r".at","the bat went splat")
```

```python
re.findall(r"....at","the bat went splat") # the more "." shows more characters
```

```python
# one or more whitespace that ends with at
re.findall(r"\S+at","the bat went splat at the bat") # "S" for non white spaces
```

```python
re.findall(r"\s+at","the bat went splat at the bat") # "s" for white spaces
```

#**Start with And Ends with**

```python
# starts with a number

re.findall(r'^\d','1 is neither prime nor composite as a number') # "^" to search for a character that starts whith
```

```python
# ends with a number

re.findall(r'\d$','the first even number is 2') # "$" for a ending character
#
```

```python
'''to exclude character, we can use '^' symbol in conjuction with a set of brackets []
   anything inside the bracket is exclded'''

phrase = "there are 3 numbers 34 inside 5 sentencaes"
re.findall(r'[^\d]',phrase)
```

```python
re.findall(r'[^\d]+',phrase)
```

```python
test_phrase = 'this is a string! but it has punctuation. how can we remove it?'
```

```python
re.findall('[^!.? ]+', test_phrase) # makes it exclude the space
```

```python
re.findall('[^!.? ]+ ', test_phrase) # makes it to tkoenize every word except puntion words
```

```python
re.findall('[^!.?]+', test_phrase)
```

```python
clean = ' '.join(re.findall('[^!.?]+ ', test_phrase))
clean
```

```python
clean = ' '.join(re.findall('[^!.?]+ ', test_phrase)) # adding space after '+' makes the whole word counts as a exception
clean
```

```python
clean = ' '.join(re.findall('[^!.?]+  ', test_phrase)) # adding double space makes it count the whole token as exception
clean
```

```python
re.findall('[^!.?]+ ', test_phrase)
```

```python
re.findall(r'[\w]+-[\w]+' , text)  # Find only hyphenated words
```
