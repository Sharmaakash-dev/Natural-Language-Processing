```python
import nltk
from nltk.stem import PorterStemmer
from nltk.stem.snowball import SnowballStemmer
```

```python
words = ['charging' , 'charges' , 'charged' , 'charge' , 'study' , 'studying' , 'easily' , 'fairly' , 'losses' , 'was' , 'necessary' , 'implementation' , 'permission' , 'learn']
```

```python
for word in words:
    print(word + ' : ' + PorterStemmer().stem(word))
```

```python
sent = "police is chasing the suspect as it ran away".split()
```

```python
sent
```

```python
for token in sent:
  print(PorterStemmer().stem(token))
```

#**SNOWBALL STREMMER**

```python
snow_stremmer = SnowballStemmer(language="english")
```

```python
for word in sent:
  print(snow_stremmer.stem(word))
```
